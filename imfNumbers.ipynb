```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
nBins=1000
ms=10**np.linspace(-4,4,nBins)
epsDm=0.158*np.exp(-np.power(np.log10(ms) - np.log10(0.08),2) / (2*(0.69**2)))
```

```python
eps=np.zeros(nBins-2)
for i in range(nBins-2):
    eps[i]=2*epsDm[i]/(ms[i]+ms[i+2]) #in theory in Msun pc-3
```

```python
plt.semilogx(ms,epsDm)
plt.show()
```

```python
plt.loglog(ms[1:-1],eps)
plt.show()
```

```python
data=np.vstack((ms[1:-1],eps)).T
print(data.shape)
```

```python
np.savetxt('data/chabrierImf.txt',data,fmt='%1.3e \t',header='M (M_s) \t dN/dMdV (M_s^-1 pc^-3)')
```
